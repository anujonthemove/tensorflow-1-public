```python
with open('lrs.pickle', 'rb') as handle:
    lrs = pickle.load(handle)

with open('history.pickle', 'rb') as handle:
    history = pickle.load(handle)
```

```python
# Define the learning rate array
# lrs = 1e-8 * (10 ** (np.arange(100) / 20))

# Set the figure size
plt.figure(figsize=(10, 6))

# Set the grid
plt.grid(True)

# Plot the loss in log scale
# plt.semilogx(lrs, history.history["loss"])
plt.semilogx(lrs, history)

# Increase the tickmarks size
plt.tick_params('both', length=10, width=1, which='both')

# Set the plot boundaries
plt.axis([1e-8, 1e-3, 0, 300])
```

```python
plt.figure(figsize=(15, 8))
plt.semilogx(lrs, history)
# Set the grid
# Increase the tickmarks size
plt.tick_params('both', length=10, width=1, which='both')

plt.grid(True)
plt.axis([1e-8, 1e-3, 0, 300])
```

```python
"""
if you take log of x and y values and then plot then by the looks of it this graph is same as the one above
but it gets difficult to then pin point which value should we take. so in that case, we will have to take
the exponent of the log which is tedious. So we use semilogx which will convert the axis to log without
changing the values, then we can pin point to the value directly

"""
plt.figure(figsize=(15, 8))
plt.plot(np.log(lrs), np.log(history))
# Set the grid
# Increase the tickmarks size
plt.tick_params('both', length=10, width=1, which='both')

plt.grid(True)
# plt.axis([1e-8, 1e-3, 0, 300])
```
